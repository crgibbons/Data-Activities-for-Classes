# Hugging Face Quickstart (Codespaces)

This notebook verifies that the Codespaces environment is working and demonstrates:

- loading a tokenizer
- loading a pretrained BERT-style model
- running inference


```python
import sys
print(sys.version)
```

## Import Transformers

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

## Load a pretrained model

```python
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

## Run inference on one sentence

```python
text = "Decisions by Design is a weird class (complimentary)." # try adding "(complimentary)" to the end

inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)

logits = outputs.logits
predicted_class_id = logits.argmax().item()
predicted_class_id
```

## Convert prediction to label

```python
label = model.config.id2label[predicted_class_id]
label
```

## Batch prediction example

```python
texts = [
    "This class is challenging.", # try adding "(derogatory)" to the end
    "Professor Gibbons is fine but evil.",
    "Professor Gibbons is evil but fine.", # try adding "(complimentary)" to the end
]

inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
outputs = model(**inputs)

predictions = torch.argmax(outputs.logits, dim=1)

[model.config.id2label[p.item()] for p in predictions]
```

## Compare with pipeline shortcut

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis", model=model_name)
classifier(texts)
```
