# NESCAC Ice Hockey (M) Principal Component Analysis
## 2022-23 through 2025-26 standings and team stats

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt

import pandas as pd
import glob
```

```python
# Find all CSV files in the current folder
csv_files = glob.glob('*.csv')

# Assuming there are exactly 4 CSV files, load them into DataFrames and drop 'teams' column
df1 = pd.read_csv(csv_files[0]).drop(columns=['Team','Neutral','Neutral Att','Neutral Att Avg','All Att','All Att Avg','All'])
df2 = pd.read_csv(csv_files[1]).drop(columns=['Team','Neutral','Neutral Att','Neutral Att Avg','All Att','All Att Avg','All'])
df3 = pd.read_csv(csv_files[2]).drop(columns=['Team','Neutral','Neutral Att','Neutral Att Avg','All Att','All Att Avg','All'])
df4 = pd.read_csv(csv_files[3]).drop(columns=['Team','Neutral','Neutral Att','Neutral Att Avg','All Att','All Att Avg','All'])
```

## Clean the data, clean clean the data!

```python
# Split the 'GF-GA' column into 'GF' and 'GA' for each dataframe
for df in [df1, df2, df3, df4]:
    df[['GF', 'GA']] = df['GF-GA'].str.split(' - ', expand=True).astype(int)
    df.drop(columns=['GF-GA'], inplace=True)
for df in [df1, df2, df3, df4]:
    df[['AVGH', 'AVGV']] = df['AVG Score'].str.split(' - ', expand=True).astype(float)
    df.drop(columns=['AVG Score'], inplace=True)
for df in [df1, df2, df3, df4]:
    df[['Pen Num', 'Pen Min']] = df['Pen'].str.split(' - ', expand=True).astype(str)
    df.drop(columns=['Pen'], inplace=True)
for df in [df1, df2, df3, df4]:
    df['Pen Num'] = df['Pen Num'].str.replace('None', '0').astype(int)
    df['Pen Min'] = df['Pen Min'].str.replace(':00', '').astype(int)
```

```python
features = [df1.columns[i] for i in range(1, len(df1.columns))]
features
```

```python
# Compute sample mean for numeric columns in df1
means = df1.mean(numeric_only=True)
print("Sample Means:")
print(means)

```

```python
# Compute sample variance for numeric columns in df1
variances = df1.var(numeric_only=True)
print("\nSample Variances:")
print(variances)
```

```python
# Compute the covariance matrix for df1
cov_matrix = df1.cov()
print("Covariance Matrix:")
print(cov_matrix)
```

```python
pca = PCA(n_components=3)
scores = pca.fit_transform(df1[features])

pca.explained_variance_ratio_
```

```python
loadings = pd.DataFrame(
    pca.components_.T,
    index=features,
    columns=["PC1","PC2","PC3"]
)

loadings
```

```python
df1['champion'] = (df1['Standing'] == 1)
```

```python
plt.scatter(scores[:,0], scores[:,1], c=df1['Standing'], cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
plt.scatter(scores[:,0], scores[:,1], c=df1['Standing'], cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")

# Add loading vectors
for i, feature in enumerate(features):
    plt.arrow(0, 0, loadings.iloc[i, 0]*3, loadings.iloc[i, 1]*3, 
              head_width=0.1, head_length=0.1, fc='red', ec='red', alpha=0.7)
    plt.text(loadings.iloc[i, 0]*3.2, loadings.iloc[i, 1]*3.2, feature, 
             fontsize=8, ha='center', va='center')

plt.colorbar(label='Standing')
plt.show()
```
