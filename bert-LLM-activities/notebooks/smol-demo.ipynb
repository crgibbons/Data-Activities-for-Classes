# SmolLM Text Generation Quickstart

This notebook demonstrates synthetic text generation using a lightweight instruction-tuned language model:

**HuggingFaceTB/SmolLM2-135M-Instruct**

We will:

- load the tokenizer and model
- generate text from prompts
- experiment with sampling parameters

```python
import os
from huggingface_hub import login

login(token=os.environ["HF_TOKEN"])
```

## Import libraries

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

## Load model and tokenizer

The first time this runs, the model will download.

```python
model_name = "HuggingFaceTB/SmolLM2-135M-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32
)
```

## Define a helper function for text generation

```python
def generate_text(prompt, max_tokens=200, temperature=0.7):
    messages = [{"role": "user", "content": prompt}]

    formatted = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer([formatted], return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=temperature,
        do_sample=True
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

## Example: synthetic news article

```python
prompt = "Write a short fictional newspaper article about a mathematician discovering a paradox in voting systems."

print(generate_text(prompt))
```

## Example: mathematical exposition

```python
prompt = "Explain in simple terms what an eigenvector is and why it matters."

print(generate_text(prompt))
```

## Experiment with temperature

Higher temperature → more creative output

Lower temperature → more deterministic output

```python
prompt = "Invent a new theorem about prime numbers."

print(generate_text(prompt, temperature=10.0))
```

## Try your own prompt

```python
prompt = "Describe Ayn Rand's philosophy of Objectivism in a single sentence."

print(generate_text(prompt, temperature=100.0))
```
